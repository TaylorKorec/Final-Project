## 2.) Fixing the Data types
### We will first remove the dollar sign and ','
### Then change the data type to decimal because this data is in US dollars

```python
# create a function to fix the rows with both $ and ,
def fixing_with_both(value):
    return Decimal(value.replace('$', '').replace(',', '').strip()) # clean row and convert to decimal

# create list of the cols and rows we want to fix
cols_to_fix = ['all_consumer', 'first', 'second', 'third', 'fourth', 'fifth']
rows_with_both = [2, 15]

# loop through each of the rows
for col in cols_to_fix:
    for row in rows_with_both:
        df.loc[row, col] = fixing_with_both(df.loc[row, col])
```

```python
# check the data type of one row's values
df.loc[2, cols_to_fix].apply(type)
```

```python
# create a function to fix the rows with just a ,
def fixing_with_one(value):
    return Decimal(value.replace(',', '').strip())
    
cols_to_fix = ['all_consumer', 'first', 'second', 'third', 'fourth', 'fifth']
rows_with_one = [0, 6]+ list(range(20, 146, 5)) #pick specfic rows + a range or rows
# we will need the data from row 20 - 145 in increments of 5

# loop through all the rows
for col in cols_to_fix:
    for row in rows_with_one:
        df.loc[row, col] = fixing_with_one(df.loc[row, col])
```

# Section 3: Data Analysis and Visualization

## Question 1: "Out of the five quintiles if income, which quintile spends the most on fresh produce?"

### First we will create a new df with a new column of the fresh produce rows, by select the produce rows in the 'item' column
### Then for each quintile column we will sum each produce row get the total spent
### Then find the group that spent the highest amount

```python
# create a list of the quintile cols
quintile_cols = ['first', 'second', 'third', 'fourth', 'fifth']

# create a list of the produce rows 
# leaving out the 'mean_' to select after
produce_rows = ['fruits_vegetables', 'fresh_fruits', 'fresh_vegetables']

# create a new df with only rows in the item col that start with 'mean_'
mean_produce_rows = df[df['item'].str.startswith('mean_')].copy()

# from the new item col, create a new col fresh_produce
# then remove the 'mean_' by replacing with an empty string ''
mean_produce_rows['fresh_produce'] = mean_produce_rows['item'].str.replace('mean_', '')

# now use .isin() for the new fresh_produce col to only keep the rows that are in the list
df_fresh_produce = mean_produce_rows[mean_produce_rows['fresh_produce'].isin(produce_rows)]

# drop the item col 
# now set the fresh_produce col as the index to label each row
df_fresh_produce = df_fresh_produce.drop(columns='item').set_index('fresh_produce')

# remove the all_consumer col thats not needed
df_fresh_produce= df_fresh_produce.drop(columns= 'all_consumer')

# calculate the sum of the fresh produce values for each quintile in a series
sum_produce = df_fresh_produce.sum()

# print the results
print("Total spent on fresh produce")
print("-----------------------------------")
# loop in each col and print the value of the col
for col, total in sum_produce.items():
    print(f"The {col} quintile spent a total of: {total}")

# use .idxmax() to get the quintile that spent the most
highest_produce = sum_produce.idxmax()
# use .max() to get the value of that quintile 
highest_produce_val = sum_produce.max()

print("---------------------------------")
print(f"The {highest_produce} income quintile spent the most on fresh produce with a total of: {highest_produce_val}$")
```

## Question 1: Data Visualization Bar plot

```python
# convert to a df with sum_produce items as the rows
df_fresh_produce = pd.DataFrame(list(sum_produce.items()), columns=['quintile','total'])
df_fresh_produce.head()
```

```python
# sizing the the plot
plt.figure(figsize=(9,5))

# create the barplot
produce_plot= sns.barplot(data=df_fresh_produce, x='quintile', y= 'total', width=0.6, color='seagreen')

# add the total to the top of each bar

# get the height of each bar in the plot
bars = produce_plot.containers[0]
heights = [bar.get_height() for bar in bars] 

# add the text to each plot
for bar, height in zip(bars, heights):
    produce_plot.text(
        bar.get_x() + bar.get_width()/2.,
        height,
        f'${height:.2f}',
        ha='center', 
        va='bottom'
    )

# label the plot
plt.title('Total Spent on Fresh Produce')
plt.ylabel('Total Spent (')
plt.xlabel('Per Quintile')

# add grid lines
plt.grid(axis='y')

plt.savefig("ProducePlot.png")

plt.show()
```

## Question 2:
### "Which quintile spent the most on processed foods?"

```python
# create a list of the processed food rows
processed_foods = ['cereals_bakery_products', 'cereal_products', 'bakery_products', 'processed_fruits_vegetables', 'sugar_other_sweets']

# create a new df with only rows in the item col that start with 'mean_'
mean_processed_rows = df[df['item'].str.startswith('mean_')].copy()

# from the new item col, create a new col for the processed foods
# remove the 'mean_' by replacing with an empty string ''
mean_processed_rows['processed_foods'] = mean_processed_rows['item'].str.replace('mean_', '')

# now use .isin() for the new col and only keep the rows that are in the list
df_processed_foods = mean_processed_rows[mean_processed_rows['processed_foods'].isin(processed_foods)]

# drop the item col and set the processed foods col as the index
df_processed_foods = df_processed_foods.drop(columns='item').set_index('processed_foods')

# remove the all_consumer col thats not needed
df_processed_foods= df_processed_foods.drop(columns= 'all_consumer')

# calculate the sum for each quintile to get the total spent on processed foods
sum_processed = df_processed_foods.sum()

# print the results
print("Total spent on Processed Foods")
print("-----------------------------------")

# loop in each col and print the value of the col
for col, total in sum_processed.items():
    print(f"The {col} quintile spent a total of: {total}$")

# use .idxmax() to get the quintile that spent the most
highest_processed = sum_processed.idxmax()
# use .max() to get the value of that quintile 
highest_processed_val = sum_processed.max()

print("---------------------------------")
print(f"The {highest_processed} income quintile spent the most on processed foods: {highest_processed_val}$")
```

## Question 2: Data Visualization Bar plot

```python
# covert the results dict to a df for viz
df_processed_foods= pd.DataFrame(list(sum_processed.items()), columns=['quintile','total'])

# size the the plot
plt.figure(figsize=(9, 5))

# create the bar plot
processed_plot= sns.barplot(data=df_processed_foods, x='quintile', y= 'total', width=0.6, color='salmon')

# add the total to the top of each bar

# get the height of each bar in the plot
bars = processed_plot.containers[0]
heights = [bar.get_height() for bar in bars] 

# add the text that is the total spent to each bar 
for bar, height in zip(bars, heights):
    processed_plot.text(
        bar.get_x() + bar.get_width()/2.,
        height,
        f'${height:.2f}',
        ha='center', 
        va='bottom'
    )

# label each part of plot 
plt.title('Total Spent on Processed Foods')
plt.ylabel('Total Spent')
plt.xlabel('Per Quintile')

plt.savefig("ProcessedPlot.png")

# add grid lines
plt.grid(axis='y')
plt.show()
```

## Question 3: What were the top food categories that consumers spent the most on for each quintile of income?

```python
# Create a list of each column
quintile_cols = ['first', 'second', 'third', 'fourth', 'fifth']

# Create a list of each food category with the food rows
processed_foods = ['cereals_bakery_products', 'cereal_products', 'bakery_products', 'processed_fruits_vegetables', 'sugar_other_sweets']
fresh_produce = ['fruits_vegetables', 'fresh_fruits', 'fresh_vegetables']
meats_seafood = ['beef', 'pork', 'other_meats', 'poultry', 'meats_poultry_fish_eggs', 'fish_seafood']
dairy_products = ['eggs', 'dairy_products', 'milk_cream', 'other_dairy_products']
misc_foods = ['fats_oils', 'miscellaneous_foods']
food_away_from_home = ['food_prepared_outoftown', 'food_away_from_home']


# create a new df with only rows in the item col that start with 'mean_'
mean_cat_col = df[df['item'].str.startswith('mean_')].copy()

# using new item col create a new col for each food category
# remove the 'mean_' by replacing with an empty string ''
mean_cat_col['category'] = mean_cat_col['item'].str.replace('mean_', '')

# select rows that only exist in the category lists
df_food = mean_cat_col[mean_cat_col['category'].isin(processed_foods + fresh_produce + meats_seafood + dairy_products + misc_foods + food_away_from_home)]

# drop the item col and set the category col as the index
df_food = df_food.drop(columns='item').set_index('category')

# drop the all consumer col
df_food= df_food.drop(columns= 'all_consumer')

df_food
```

```python
# calculate the total spent for each catagory by the lists 
sum_processed_foods= df_food[df_food.index.isin(processed_foods)].sum()
sum_fresh_produce= df_food[df_food.index.isin(fresh_produce)].sum()
sum_meats_seafood= df_food[df_food.index.isin(meats_seafood)].sum()
sum_dairy_products=df_food[df_food.index.isin(dairy_products)].sum()
sum_misc_foods=df_food[df_food.index.isin(misc_foods)].sum()
sum_food_away_from_home=df_food[df_food.index.isin(food_away_from_home)].sum()
```

```python
# place the sum into a dict
category_totals = {
    'processed_foods': sum_processed_foods,
    'fresh_produce': sum_fresh_produce,
    'meats_seafood': sum_meats_seafood,
    'dairy_products': sum_dairy_products,
    'misc_foods': sum_misc_foods,
    'food_away_from_home': sum_food_away_from_home
}
```

```python
# convert the dict to a df
df_cat_total = pd.DataFrame(category_totals)
df_cat_total.head()
```

```python
# find the highest col and value in each row
highest_categories = df_cat_total.idxmax(axis=1)
highest_values = df_cat_total.max(axis=1)

# loop in each quintile 
for quintile in df_cat_total.index:
    highest_cat = highest_categories[quintile]
    highest_cat_val = highest_values[quintile]
    print(f"The {quintile} quintile spent the most on '{highest_cat}' with a total of: {highest_cat_val}$")
```

```python
# check the datatype of the values in the df
df_cat_total.dtypes
```

```python
# change to a numeric datatype 
df_cat_total= df_cat_total.apply(pd.to_numeric)
df_cat_total.dtypes
```

```python
# create a stacked bar plot to show the spending in each catagory

# set the colors
color = ['skyblue', 'seagreen', 'wheat', 'violet', 'mediumpurple', 'lightcoral']

# create the stacked bar plot
df_cat_total.plot(kind='bar', stacked=True, figsize=(10, 7), width=0.6, color=color)


# label the title and axis's and set font
plt.title('Spending in Each Food Category Per Quintile', size=16)
plt.xlabel("Quintiles", size= 15)
plt.ylabel("Total Spending", size= 15)

# rotate the x-axis labels
plt.xticks(rotation=30) 

# create legend
plt.legend(title="Food Categories")

plt.savefig("CategoryPlot.png")

# display the stacked bar plot
plt.show()
```

## Question 4: Is there a significant difference in the spending on each food category between each quintiles?

```python
# move the quinties as the index into a col
plot_df = df_cat_total.reset_index(names='quantile')
# change the col to a string datatype
plot_df['quantile'] = plot_df['quantile'].astype(str)
```

### Create a lineplot to show each line as a food catagory
